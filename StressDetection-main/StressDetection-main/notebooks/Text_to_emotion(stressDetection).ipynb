```python
from transformers import pipeline

# Load emotion classifier
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", top_k=1)

# Input prompt
text = "What the hell is happening here !"

# Detect emotion
emotion = classifier(text)
print(emotion)
```

```python
from transformers import pipeline

# Load the model
model_name = "j-hartmann/emotion-english-distilroberta-base"
classifier = pipeline("text-classification", model=model_name, top_k=1)

# Save the model and tokenizer
from transformers import AutoModelForSequenceClassification, AutoTokenizer

model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Save locally (creates a folder with all files)
model.save_pretrained("./emotion-model")
tokenizer.save_pretrained("./emotion-model")
```
